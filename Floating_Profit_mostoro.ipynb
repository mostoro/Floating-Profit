# Importings

```python
import pandas as pd
from arctic import Arctic
from arctic.date import DateRange
from arctic.hooks import register_get_auth_hook
from arctic.auth import Credential
import seaborn as sns
import pandas_bokeh
import datetime
```

# Loading data

```python
def arctic_auth_hook(mongo_host, app, database):
    return Credential(database='arctic', user='moises',password='spartan300',)

start = '22/02/2021 13:00:00' #start date of trades dataframe
end = '04/18/2021 18:00:00' #end date of trades dataframe

register_get_auth_hook(arctic_auth_hook)
store = Arctic('157.245.223.103')
library = store['Futures_Historical_Ticks']
item = library.read('MNQ', date_range=DateRange(start, end))
data = item.data
data = data['Last'].resample('5S').ohlc()
```

# Floating Profit

Entry variables for analize each trade

```python
tick_value = 0.5
tick_size = 0.25
entry_date = '04/01/2021 09:37:30'
exit_date = '04/01/2021 10:54:39'
entry_price = 13285.00
exit_price = 13288.25
long_short = 'LONG'
```

Plotting the performance of each trade

```python
floating_ticks = pd.DataFrame(columns=['date', 'floating_profit_ticks_cum'])
#floating_usd = pd.DataFrame(columns=['date', 'floating_profit_usd_cum'])


for index, row in data[entry_date:exit_date].iterrows():
    print(index, 'close:', row['close'])
    if long_short == 'LONG':
        floating_profit_ticks = row['close'] - entry_price
        floating_ticks = floating_ticks.append({'date': index, 'floating_profit_ticks_cum': floating_profit_ticks},ignore_index=True)

        #floating_profit_usd = floating_profit_ticks * tick_value
        #floating_usd = floating_usd.append({'date': index, 'floating_profit_usd_cum': floating_profit_usd},ignore_index=True)
        
        print(floating_profit_ticks, 'ticks')
        #print(floating_profit_usd, 'usd')
    else:
        floating_profit_ticks = entry_price - row['close']
        floating_ticks = floating_ticks.append({'date': index, 'floating_profit_ticks_cum': floating_profit_ticks},ignore_index=True)

        #floating_profit_usd = floating_profit_ticks * tick_value
        #floating_usd = floating_usd.append({'date': index, 'floating_profit_usd_cum': floating_profit_usd},ignore_index=True)
    

        print(floating_profit_ticks, 'ticks') 
        #print(floating_profit_usd, 'usd')
    
pd.set_option('plotting.backend', 'pandas_bokeh')
#pandas_bokeh.output_notebook()
pandas_bokeh.output_file('Interactive Plot.html')
floating_ticks.dropna().set_index('date').plot_bokeh(kind= 'line')
#floating_usd.set_index('date').plot_bokeh(kind= 'line')
```
